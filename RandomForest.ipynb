```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data.data
```

```python
data.feature_names
```

```python
data.target
```

```python
data.target_names
```

```python
df = pd.DataFrame(np.c_[data.data,data.target],columns=[list(data.feature_names) + ["target"]])
df.head()
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2020)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=100,criterion="gini")
classifier.fit(X_train,y_train)
```

```python
classifier.score(X_test,y_test)
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/housepriceprediction/bangalore house price prediction OHE-data.csv")
df.head()
```

```python
X = df.drop("price",axis=1)
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2020)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=100,criterion="squared_error")  #  these parameters are default parameters

regressor.fit(X_train,y_train)
```

```python
regressor.score(X_test,y_test)
```
