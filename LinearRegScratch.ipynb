```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/teams (1).csv")
df.head()
```

```python
df.shape
```

```python
df.info()      # no null value available
```

```python
X1 = df[['athletes','prev_medals']].copy()
Y = df[['medals']].copy()
```

```python
X1,Y
```

```python
X1['intercept'] = 1
X2 = X1[['intercept','athletes','prev_medals']]
```

```python
X2
```

```python
X2_transpose = X2.T
X2_transpose
```

```python
B = np.linalg.inv(X2_transpose @ X2) @ X2_transpose @ Y       # result is b0 which is intercept,b1,b2
B.index = X2.columns
B
```

```python
predictions = X2 @ B
predictions
```

```python
SSR = ((Y - predictions)**2).sum()
SSR
```

```python
SST = ((Y - Y.mean())**2).sum()
SST
```

```python
R_square = (1 - (SSR/SST))
R_square         # R2 value is 0.87 which is close to 1 means prediction is good.
```
