```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data.data
```

```python
data.feature_names
```

```python
data.target
```

```python
data.target_names
```

```python
df = pd.DataFrame(np.c_[data.data,data.target],columns=[list(data.feature_names) + ["target"]])
df.head()
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2020)
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5) # this 5 is default value and no neeed to give this
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.score(X_test,y_test)
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/housepriceprediction/bangalore house price prediction OHE-data.csv")
df.head()
```

```python
X = df.drop("price",axis=1)
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2020)
```

```python
from sklearn.neighbors import KNeighborsRegressor

regressor = KNeighborsRegressor(n_neighbors=5)
```

```python
regressor.fit(X_train,y_train)
```

```python
regressor.score(X_test,y_test)
```

```python
regressor.predict([X_test.iloc[-1,:]])
```
